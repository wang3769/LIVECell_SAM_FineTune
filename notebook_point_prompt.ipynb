```python
import json
import torch
from torch.utils.data import Dataset
import numpy as np
from PIL import Image
from pycocotools.coco import COCO
from segment_anything.utils.transforms import ResizeLongestSide


class LiveCellSAMDataset(Dataset):
    def __init__(self, img_dir, ann_file, sam_model):
        self.img_dir = img_dir
        self.coco = COCO(ann_file)
        self.image_ids = list(self.coco.imgs.keys())

        # SAM transform (fixed contract)
        self.transform = ResizeLongestSide(1024)

        # required for normalization
        self.pixel_mean = sam_model.pixel_mean
        self.pixel_std = sam_model.pixel_std

    def __len__(self):
        return len(self.image_ids)

    def _sample_point(self, mask):
        ys, xs = torch.where(mask > 0)
        idx = torch.randint(0, len(xs), (1,))
        return torch.stack([xs[idx], ys[idx]], dim=-1)

    def __getitem__(self, idx):
        image_id = self.image_ids[idx]
        img_info = self.coco.imgs[image_id]

        # ---- load image ----
        image = Image.open(f"{self.img_dir}/{img_info['file_name']}").convert("RGB")
        image = np.array(image)

        H, W = image.shape[:2]

        # ---- load instance masks ----
        ann_ids = self.coco.getAnnIds(imgIds=image_id)
        anns = self.coco.loadAnns(ann_ids)

        masks = []
        for ann in anns:
            masks.append(self.coco.annToMask(ann))

        masks = torch.from_numpy(np.stack(masks)).to(torch.uint8)  # [N,H,W]

        # ---- pick ONE instance ----
        inst_idx = torch.randint(0, masks.shape[0], (1,)).item()
        gt_mask = masks[inst_idx]                                   # [H,W]

        # ---- sample prompt ----
        point = self._sample_point(gt_mask)                         # [1,2]

        # ---- apply SAM transforms ----
        image = self.transform.apply_image(image)
        gt_mask = self.transform.apply_image(gt_mask.numpy())
        point = self.transform.apply_coords(point.numpy(), (H, W))

        # ---- tensorize & normalize ----
        image = torch.from_numpy(image).permute(2, 0, 1).float()
        image = image / 255.0

        gt_mask = torch.from_numpy(gt_mask).unsqueeze(0).float()
        point = torch.from_numpy(point).float()

        return {
            "image": image,
            "mask": gt_mask,
            "point": point,
            "image_id": image_id,
        }
```

```python
import json
import torch
from torch.utils.data import Dataset
import numpy as np
from PIL import Image
from pycocotools.coco import COCO
from segment_anything.utils.transforms import ResizeLongestSide


class LiveCellSAMDataset(Dataset):
    def __init__(self, img_dir, ann_file, sam_model):
        self.img_dir = img_dir
        self.coco = COCO(ann_file)
        self.image_ids = list(self.coco.imgs.keys())

        # SAM resize contract
        self.transform = ResizeLongestSide(1024)

        # normalization (buffers on CPU here)
        self.pixel_mean = sam_model.pixel_mean
        self.pixel_std = sam_model.pixel_std

    def __len__(self):
        return len(self.image_ids)

    def _sample_point(self, mask): # yes, the sampling is from the gt_mask, but image this is like pick a point by human
        ys, xs = torch.where(mask > 0)

        # robust fallback for empty masks
        if len(xs) == 0:
            H, W = mask.shape
            x = torch.randint(0, W, (1,))
            y = torch.randint(0, H, (1,))
            return torch.stack([x, y], dim=-1)

        idx = torch.randint(0, len(xs), (1,))
        return torch.stack([xs[idx], ys[idx]], dim=-1)

    def __getitem__(self, idx):
        image_id = self.image_ids[idx]
        img_info = self.coco.imgs[image_id]

        # ---- load image ----
        image = Image.open(f"{self.img_dir}/{img_info['file_name']}").convert("RGB")
        image = np.array(image)
        H, W = image.shape[:2]

        # ---- load masks ----
        ann_ids = self.coco.getAnnIds(imgIds=image_id)
        anns = self.coco.loadAnns(ann_ids)

        masks = []
        for ann in anns:
            masks.append(self.coco.annToMask(ann))

        masks = torch.from_numpy(np.stack(masks)).to(torch.uint8)  # [N,H,W]

        # ---- pick ONE instance ----
        inst_idx = torch.randint(0, masks.shape[0], (1,)).item()
        gt_mask = masks[inst_idx]                                  # [H,W]

        # ---- sample prompt ----
        point = self._sample_point(gt_mask)                        # [1,2]

        # ---- SAM resize (aspect-ratio preserved) ----
        image = self.transform.apply_image(image)
        gt_mask = self.transform.apply_image(gt_mask.numpy())
        point = self.transform.apply_coords(point.numpy(), (H, W))

        # ---- padding to 10241024 ----
        h, w = image.shape[:2]
        pad_h = 1024 - h
        pad_w = 1024 - w

        image = np.pad(
            image,
            ((0, pad_h), (0, pad_w), (0, 0)),
            mode="constant",
            constant_values=0,
        )

        gt_mask = np.pad(
            gt_mask,
            ((0, pad_h), (0, pad_w)),
            mode="constant",
            constant_values=0,
        )

        # ---- tensorize ----
        image = torch.from_numpy(image).permute(2, 0, 1).float() / 255.0
        gt_mask = torch.from_numpy(gt_mask).unsqueeze(0).float()
        point = torch.from_numpy(point).float()

        return {
            "image": image,    # [3,1024,1024]
            "mask": gt_mask,   # [1,1024,1024]
            "point": point,    # [1,2]
            "image_id": image_id,
        }
```

```python
from torch.utils.data import DataLoader
from tqdm import tqdm


# freeze everything except mask decoder
for p in sam.image_encoder.parameters():
    p.requires_grad = False
for p in sam.prompt_encoder.parameters():
    p.requires_grad = False

optimizer = torch.optim.AdamW(
    sam.mask_decoder.parameters(), lr=1e-4
)

loader = DataLoader(dataset, batch_size=1, shuffle=True) # very useful not just for convenience
```

```python
sam.train()

for batch in loader:
    # ---- move to device ----
    img = batch["image"].to(device)      # [B,3,1024,1024]
    gt_mask = batch["mask"].to(device)   # [B,1,1024,1024]
    point = batch["point"].to(device)    # [B,1,2]

    # ---- normalize for SAM ----
    img = (img - sam.pixel_mean) / sam.pixel_std

    # ---- image encoder ----
    with torch.no_grad():
        image_embedding = sam.image_encoder(img)  # [B,256,64,64]

    # ---- prompt encoder ----
    labels = torch.ones(
        (point.shape[0], 1),
        device=device,
        dtype=torch.int64
    )

    sparse, dense = sam.prompt_encoder(
        points=(point, labels),
        boxes=None,
        masks=None
    )
    # sparse: [B,2,256]  (EXPECTED)
    # dense : [B,256,64,64]

    # ---- mask decoder (POSITIONAL args only) ----
    low_res_masks, _ = sam.mask_decoder(
        image_embedding,
        sam.prompt_encoder.get_dense_pe().to(device),
        sparse,
        dense,
        multimask_output=False
    )
    # low_res_masks: [B,1,256,256]

    print("OK forward pass:", low_res_masks.shape)
    break
```

```python
# why 256 x 256? because Image encoder outputs features at 1/16 resolution
# yet the output is just logits for cross entropy convenience we need to look at prob
mask_prob = torch.sigmoid(low_res_masks)
print(mask_prob)
```

```python
import torch.nn.functional as F

# low_res_masks: [B, 1, 256, 256]
pred_logits = low_res_masks

# probability mask
pred_prob = torch.sigmoid(pred_logits)

# upsample to image size (1024x1024)
pred_prob_up = F.interpolate(
    pred_prob,
    size=(1024, 1024),
    mode="bilinear",
    align_corners=False
)

# binary mask (for evaluation / overlay)
pred_mask = (pred_prob_up > 0.5).float()
```

```python
# flatten for safety
pred_flat = pred_mask.view(-1)
gt_flat = gt_mask.view(-1)

intersection = (pred_flat * gt_flat).sum()
union = pred_flat.sum() + gt_flat.sum() - intersection

iou = intersection / (union + 1e-6)
dice = 2 * intersection / (pred_flat.sum() + gt_flat.sum() + 1e-6)

print(f"IoU  : {iou.item():.4f}")
print(f"Dice : {dice.item():.4f}")
```

```python
def iou_loss(logits, targets, eps=1e-6):
    """
    logits: [B,1,H,W] raw SAM outputs
    targets: [B,1,H,W] binary {0,1}
    """
    probs = torch.sigmoid(logits)

    intersection = (probs * targets).sum(dim=(2, 3))
    union = probs.sum(dim=(2, 3)) + targets.sum(dim=(2, 3)) - intersection

    iou = (intersection + eps) / (union + eps)
    loss = 1 - iou

    return loss.mean()
```

```python
# need to change the ground truth gt_mask dimension, downsize using 
gt_mask_256 = torch.nn.functional.interpolate(
    gt_mask,
    size=(256, 256),
    mode="nearest"
)
```

```python
sam.train()

num_epochs = 1
loss_history = []
global_step = 0

for epoch in range(num_epochs):
    for batch_idx, batch in enumerate(loader):
        img = batch["image"].to(device)      # [B,3,1024,1024]
        gt_mask = batch["mask"].to(device)   # [B,1,1024,1024]
        point = batch["point"].to(device)    # [B,1,2]

        # normalize
        img = (img - sam.pixel_mean) / sam.pixel_std

        with torch.no_grad():
            image_embedding = sam.image_encoder(img)

        labels = torch.ones(
            (point.shape[0], 1),
            device=device,
            dtype=torch.int64
        )

        sparse, dense = sam.prompt_encoder(
            points=(point, labels),
            boxes=None,
            masks=None
        )

        low_res_masks, _ = sam.mask_decoder(
            image_embedding,
            sam.prompt_encoder.get_dense_pe().to(device),
            sparse,
            dense,
            multimask_output=False
        )

        # resize GT mask to 256x256
        gt_mask_256 = torch.nn.functional.interpolate(
            gt_mask,
            size=(256, 256),
            mode="nearest"
        )

        loss = iou_loss(low_res_masks, gt_mask_256)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # ---- logging ----
        loss_history.append(loss.item())
        global_step += 1

        if global_step % 100 == 0:
            print(
                f"epoch {epoch} | step {global_step} | "
                f"batch {batch_idx} | loss {loss.item():.4f}"
            )
```
